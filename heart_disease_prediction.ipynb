```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import joblib

from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import r2_score, mean_absolute_error
```

```python
df = pd.read_csv("heart (1).csv")
print("Shape:", df.shape)
df.head()
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python

df = df.drop("Unnamed: 0",axis=1)
```

```python
d = {'cp':"chest_pain_type", 'trestbps':"resting_bp", 'chol':"cholestrol", 'fbs':"fasting_blood_sugar", 'thalach':"max_heart_rate", 'exang':"exer_induced_angina", 'ca':"num_major_vessels"}
df.rename(columns=d,inplace=True)
df.head()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import confusion_matrix,classification_report
```

```python
X = df.drop("target",axis=1)
y = df.target
```

```python
Xtrain,Xtest,ytrain,ytest = train_test_split(X,y,test_size=0.3,random_state=42)
```

```python
model = LogisticRegression()
model.fit(Xtrain,ytrain)
```

```python
print("Training Eff: ",model.score(Xtrain,ytrain))
print("Testing Eff: ",model.score(Xtest,ytest))
```

```python
ypred = model.predict(Xtest)
ypred
```

```python
confusion_matrix(ytest,ypred)
```

```python

print(classification_report(ytest,ypred))
```
